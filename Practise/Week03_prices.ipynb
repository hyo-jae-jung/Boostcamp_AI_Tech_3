```python
import numpy as np
import pandas as pd
```

```python
stock = pd.read_csv('prices.csv')
```

```python
stock.sample(10)
```

```python
stock.info()
```

```python
stock.describe(include='all')
```

```python
stock.index = stock.date
```

```python
stock = stock.iloc[:,1:]
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
apple = stock[stock['symbol']=='AAPL']
google = stock[stock['symbol']=='GOOGL']
google.head()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 7))

ax.plot(apple['close'])
ax.plot(google['close'])
ax.legend()

plt.show()
```

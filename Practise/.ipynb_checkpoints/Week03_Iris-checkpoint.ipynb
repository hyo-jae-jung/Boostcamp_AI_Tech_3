```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
iris = pd.read_csv('./iris.csv')
```

```python
iris.sample(5)
```

```python
iris.info()
```

```python
iris.describe(include='all')
```

```python
iris = pd.DataFrame(data=iris,columns=list(iris.columns[1:]),index=iris.Id)
```

```python
iris.groupby(['Species']).mean()
```

```python
sns.pairplot(data=iris, hue='Species', corner=True)
```

iris 관련 그래프는 pairplot이면 정리가 되는 것 같다.  

그래프만 봤을 때는 Species를 구분짓는 변수는 PetalLengthCm, PetalWidthCm 정도로 보인다.

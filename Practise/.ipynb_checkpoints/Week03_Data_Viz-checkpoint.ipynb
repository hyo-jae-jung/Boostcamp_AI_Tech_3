```python
fig = plt.figure()

x1 = [1,2,3]
x2 = [3,2,1]

ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

ax1.plot(x1)
ax2.plot(x2)
plt.show()
```

3-2. 색상 지정하기

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1,1,1], color="r") # 한 글자로 정하는 색상
ax.plot([2,2,2], color="forestgreen") # color name
ax.plot([3,3,3], color="#000000") # hex code 이거 추천(RGB)
plt.show()
```

3-3. 텍스트 사용하기

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([1,1,1],label='1')
ax.plot([2,2,2],label='2')
ax.plot([3,3,3],label='3')
ax.legend() # 범례 출력. label 정보를 가지고 출력한다.
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([1,1,1],label='1')
ax.plot([2,2,2],label='2')
ax.plot([3,3,3],label='3')
ax.set_title("Basic Plot") # 제목
ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.set_title('ax1') # 각각의 그래프에 제목을 달 수 있다
ax2.set_title('ax2')

fig.suptitle("fig") # 전체 제목도 달 수 있다. super
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([1,1,1],label='1')
ax.plot([2,2,2],label='2')
ax.plot([3,3,3],label='3')
ax.set_title("Basic Plot")
ax.legend()
print(ax.get_title()) # get ~ method로 다양한 정보를 불러 올 수 있다.
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([1,1,1],label='1')
ax.plot([2,2,2],label='2')
ax.plot([3,3,3],label='3')
ax.set_title("Basic Plot")
ax.set_xticks([0,1,2]) # 축에 원하는 방식으로 사용 할 수 있다.
ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([1,1,1],label='1')
ax.plot([2,2,2],label='2')
ax.plot([3,3,3],label='3')
ax.set_title("Basic Plot")
ax.set_xticks([0,1,2])
ax.set_xticklabels(['zero','one','two']) # 레이블도 추가 할 수 있다.
ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([1,1,1],label='1')
ax.plot([2,2,2],label='2')
ax.plot([3,3,3],label='3')
ax.set_title("Basic Plot")
ax.set_xticks([0,1,2])
ax.set_xticklabels(['zero','one','two'])
ax.text(x=1,y=2,s='This is Text') # 원하는 좌표에 텍스트도 입력 가능하다.
ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([1,1,1],label='1')
ax.plot([2,2,2],label='2')
ax.plot([3,3,3],label='3')
ax.set_title("Basic Plot")
ax.set_xticks([0,1,2])
ax.set_xticklabels(['zero','one','two'])
ax.annotate(text='This is Annotate', xy=(1,2))
# annotate로도 텍스트 입력 가능한데 text method와 용도가 조금 다름
ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([1,1,1],label='1')
ax.plot([2,2,2],label='2')
ax.plot([3,3,3],label='3')
ax.set_title("Basic Plot")
ax.set_xticks([0,1,2])
ax.set_xticklabels(['zero','one','two'])

ax.annotate(text='This is Annotate', xy=(1,2),
           xytext=(1.2,1.2),
           arrowprops=dict(facecolor='black'),
           ) # 이런식으로 화살표도 추가 가능. (annotate : 주석)
ax.legend()
plt.show()
```

### 2. 다양한 Bar Plot  

**2-0. 데이터 준비하기

```python
student = pd.read_csv("./StudentsPerformance.csv")
```

```python
student.sample(10)
# sample을 추천하는 이유는 순서대로 정렬된 데이터에 대해 데이터를 다양하게 볼 수 있다
```

```python
student.head()
```

```python
student.info()
```

```python
student.describe(include='all')
```

```python
group = student.groupby('gender')['race/ethnicity'].value_counts().sort_index()
display(group)
```

## 2-1. Multiple Bar Plot

```python
fig, axes = plt.subplots(1,2,figsize=(15,7))
axes[0].bar(group['male'].index, group['male'], color='royalblue')
axes[1].bar(group['female'].index, group['female'], color='tomato')
plt.show()
```

```python
fig, axes = plt.subplots(1,2,figsize=(15,7),sharey=True)
axes[0].bar(group['male'].index, group['male'], color='royalblue')
axes[1].bar(group['female'].index, group['female'], color='tomato')
plt.show()
```

```python
fig, axes = plt.subplots(1,2,figsize=(15,7))
axes[0].bar(group['male'].index, group['male'], color='royalblue')
axes[1].bar(group['female'].index, group['female'], color='tomato')

for ax in axes:
    ax.set_ylim(0,200)

plt.show()
```

### 2-2. Stacked Bar Plot

```python
fig, axes = plt.subplots(1,2,figsize=(15,7))
group_cnt = student['race/ethnicity'].value_counts().sort_index()
axes[0].bar(group_cnt.index, group_cnt, color='darkgray')
axes[1].bar(group['male'].index, group['male'], color='royalblue')
axes[1].bar(group['female'].index, group['female'], bottom=group['male'], color='tomato')

for ax in axes:
    ax.set_ylim(0,350)

plt.show()
```

female 데이터에 대한 정보를 이해하기 불편함. 가독성 나쁨

## 2-3. Percentage Stacked Bar Plot

```python
fig, ax = plt.subplots(1,1,figsize=(12,7))

group = group.sort_index(ascending=False) # 역순 정렬
total = group['male']+group['female'] # 각 그룹별 합

ax.barh(group['male'].index, group['male']/total, color='royalblue')

ax.barh(group['male'].index, group['male']/total, color='royalblue')
```
